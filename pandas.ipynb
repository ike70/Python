## INTRODUCTION TO MATPLOTLIB

```python
# Is a plotting library in python
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Create a sample dataset
x = np.linspace(0,10,100)
y = np.sin(x)

```

```python
print(x)
```

```python
# Line Plot
plt.figure(figsize=(10,4))
plt.plot(x,y, color = 'red')
plt.title('SINE WAVE', color = 'green')
plt.xlabel('X-AXIS', color = 'blue')
plt.ylabel('Y-AXIS', color = 'blue')
plt.legend(x)
plt.grid()
plt.show()
```

```python
# Bar Chart
plt.figure(figsize=(10,6))
plt.barh(data['Retailer'], data["Operating Profit"])
plt.title('PROFITS BY RETAILERS', color = 'r')
plt.xlabel('PROFITS')
plt.ylabel('RETAILERS')
plt.show()
```

```python
# PIE Chart
explode = [0.1,0.1,0.1,0.1]
fruits = ["Orange", "Mango", "Cherry", "Water Melon"]
values = [1,2,3,4]
plt.pie(values,labels = fruits, autopct = "%1.0f%%", explode = explode, shadow = True)
plt.title("Percentage of Fruits")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [25, 30, 20, 25]
colors = ['red','yellow','green','blue']

# Create pie chart
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, colors=colors, wedgeprops={'width':0.3}, startangle=90)

# Draw circle in the middle for donut shape
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')

plt.title('Fruit Distribution')
plt.show()
```
